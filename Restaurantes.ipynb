```python
# Ler objectId
from bson import ObjectId

# Abrir o arquivo JSON e carregar os dados
data = []
with open(file_path, "r", encoding='utf-8') as file:
  for line in file:
    record = json.loads(line)

    # Converter o campo '_id' com o $oid em ObjectId
    if '_id' in record and '$oid' in record['_id']:
      record['_id'] = ObjectId(record['_id']['$oid'])
    data.append(record)
```

```python
# Nome do restaurante a ser atualizado
nome_restaurante = 'A & G Kitchen'

# Lista de pratos principais a ser adicionada
pratos_principais = [
    'Pizza Margherita',
    'Lasanha Vegana',
    'Risotto de Cogumelos'
]

# Atualizando o restaurante com a lista de pratos principais
result = collection.update_one(
    {'name': nome_restaurante},  # Filtro para encontrar o restaurante
    {'$set': {'main_dishes': pratos_principais}}  # Atualiza o campo main_dishes
)

# Verificando se a atualização foi realizada
if result.matched_count > 0:
    print(f'Restaurante "{nome_restaurante}" atualizado com sucesso.')
else:
    print(f'Restaurante "{nome_restaurante}" não encontrado.')

# Buscando o restaurante atualizado para verificar
restaurante_atualizado = collection.find_one({'name': nome_restaurante})

if restaurante_atualizado:
    print("Detalhes do restaurante atualizado:")
    print(f"Nome: {restaurante_atualizado['name']}")
    print(f"Endereço: {restaurante_atualizado['address']}, {restaurante_atualizado['address line 2']}")
    print(f"Código Postal: {restaurante_atualizado['postcode']}")
    print(f"Classificação: {restaurante_atualizado['rating']}")
    print(f"Tipo de comida: {restaurante_atualizado['type_of_food']}")
    print(f"Pratos principais: {', '.join(restaurante_atualizado['main_dishes'])}")
    print(f"URL: {restaurante_atualizado['URL']}")
else:
    print("Restaurante não encontrado após a atualização.")
```
